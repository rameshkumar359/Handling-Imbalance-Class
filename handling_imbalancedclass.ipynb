#### using Ensemble technique

```python
X = df3.drop('Churn',axis='columns')
y = df3['Churn']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=15, stratify=y)
```

```python
y_train.value_counts()
```

```python
df3_class_0=df3[df3['Churn']==0]
df3_class_1=df3[df3['Churn']==1]
```

```python
df3_class_0.shape,df3_class_1.shape
```

```python
def get_train_batch(df_majority,df_minority,strat,end):
    df_train=pd.concat([df_majority[strat:end],df_minority],axis=0)
    X_train=df_train.drop('Churn',axis='columns')
    y_train=df_train['Churn']
    return X_train, y_train
```

#### batch 1

```python
X_train,y_train=get_train_batch(df3_class_0,df3_class_1,0,1869)
model=LogisticRegression()
```

```python
model.fit(X_train,y_train)
```

```python
pred_1=model.predict(X_test)
```

#### batch 2

```python
X_train,y_train=get_train_batch(df3_class_0,df3_class_1,1870,3739)
```

```python
model.fit(X_train,y_train)
```

```python
pred_2=model.predict(X_test)
```

#### batch3

```python
X_train,y_train=get_train_batch(df3_class_0,df3_class_1,3294,5163)
```

```python
model=LogisticRegression()
```

```python
model.fit(X_train,y_train)
```

```python
pred_3=model.predict(X_test)
```

```python
def majority_voting(pred_1,pred_2,pred_3):
    final_pred=pred_1.copy()
    for i in range(len(pred_1)):
        sum= pred_1[i] + pred_2[i] + pred_3[i]

        if sum > 1:
            final_pred[i]=1
        else:
            final_pred[i]=0
    return final_pred
```

```python
final_pred=majority_voting(pred_1,pred_2,pred_3)
```

```python
metrics.accuracy_score(y_test,final_pred)
```

```python
print(metrics.classification_report(y_test,final_pred))
```

#### report for ensembling technique
